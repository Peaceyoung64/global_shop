```python
%sql

CREATE OR REPLACE TABLE gold.dim_products
USING DELTA
AS

SELECT

    product_key,
    product_id,
    product_number,

    product_name,

    category,

    subcategory,

    product_line,

    maintenance,

    cost,

    CASE

        WHEN cost >= 1000 THEN 'Premium'

        WHEN cost >= 500 THEN 'Standard'

        ELSE 'Budget'

    END AS price_band,

    start_date

FROM silver.products;
```
