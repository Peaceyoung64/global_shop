```python
%sql

CREATE OR REPLACE TABLE gold.dim_date
USING DELTA
AS

WITH calendar AS (

SELECT
    EXPLODE(
        SEQUENCE(
            (SELECT MIN(order_date) FROM silver.sales),
            (SELECT MAX(order_date) FROM silver.sales),
            INTERVAL 1 DAY
        )
    ) AS date

)

SELECT

    date,

    YEAR(date) AS year,

    QUARTER(date) AS quarter,

    MONTH(date) AS month_number,

    MONTHNAME(date) AS month_name,

    CONCAT(
        YEAR(date),
        '-',
        LPAD(MONTH(date),2,'0')
    ) AS year_month,

    WEEKOFYEAR(date) AS week_number,

    DAY(date) AS day,

    DAYOFWEEK(date) AS day_of_week,

    DATE_FORMAT(date,'EEEE') AS weekday_name,

    CASE
        WHEN DAYOFWEEK(date) IN (1,7)
        THEN 'Weekend'
        ELSE 'Weekday'
    END AS day_type

FROM calendar;
```
