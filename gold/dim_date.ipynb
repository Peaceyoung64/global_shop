```python
%sql

CREATE OR REPLACE TABLE gold.dim_date
USING DELTA
AS

SELECT DISTINCT

    order_date AS date,

    YEAR(order_date) AS year,

    QUARTER(order_date) AS quarter,

    MONTH(order_date) AS month_number,

    MONTHNAME(order_date) AS month_name,

    CONCAT(

        YEAR(order_date),

        '-',

        LPAD(MONTH(order_date),2,'0')

    ) AS year_month,

    WEEKOFYEAR(order_date) AS week_number,

    DAY(order_date) AS day,

    DAYOFWEEK(order_date) AS day_of_week,

    DATE_FORMAT(order_date,'EEEE') AS weekday_name,

    CASE

        WHEN DAYOFWEEK(order_date) IN (1,7)

        THEN 'Weekend'

        ELSE 'Weekday'

    END AS day_type

FROM silver.sales;
```
