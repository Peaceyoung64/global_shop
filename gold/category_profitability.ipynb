### category_profitability

```python
%sql
CREATE OR REPLACE TABLE gold.category_profitability
USING DELTA
AS

SELECT
    category,
    subcategory,

    -- Number of unique products in each category
    COUNT(DISTINCT product_key) AS total_products,

    -- Sales volume metrics
    SUM(total_orders) AS total_orders,
    SUM(units_sold) AS units_sold,

    -- Financial metrics
    ROUND(SUM(total_revenue), 2) AS total_revenue,
    ROUND(SUM(total_profit), 2) AS total_profit,

    -- Overall profit margin for the category
    ROUND(
        SUM(total_profit) * 100.0
        / NULLIF(SUM(total_revenue), 0),
        2
    ) AS profit_margin_pct,

    -- Percentage contribution to total company revenue
    ROUND(
        SUM(total_revenue) * 100.0
        / SUM(SUM(total_revenue)) OVER (),
        2
    ) AS category_revenue_share_pct

FROM workspace.gold.product_profitability

GROUP BY
    category,
    subcategory

ORDER BY
    total_profit DESC;
```
