```python
%sql

CREATE OR REPLACE TABLE gold.dim_customers
USING DELTA
AS

SELECT

    customer_key,
    customer_id,
    customer_number,

    first_name,
    last_name,
    full_name,

    country,

    marital_status,

    gender,

    birthdate,

    age_years,

    CASE
        WHEN age_years < 18 THEN 'Under 18'
        WHEN age_years BETWEEN 18 AND 34 THEN '18 - 34'
        WHEN age_years BETWEEN 35 AND 49 THEN '35 - 49'
        WHEN age_years BETWEEN 50 AND 64 THEN '50 - 64'
        ELSE '65+'
    END AS age_group,

    create_date

FROM silver.customers;
```
