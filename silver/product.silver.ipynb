```python
%sql
CREATE OR REPLACE TABLE silver.products
USING DELTA
AS
SELECT
  product_key,
  product_id,
  product_number,
  TRIM(product_name)                    AS product_name,
  category_id,
 
  -- category: 7 rows have NULL — fill with "Accessories"
  -- (these are pedal products based on their product codes)
  COALESCE(category, 'Accessories')   AS category,
 
  -- subcategory: also NULL for same 7 rows
  COALESCE(subcategory, 'Pedals')     AS subcategory,
 
  -- maintenance: NULL rows → "Unknown"
  COALESCE(maintenance, 'Unknown')   AS maintenance,
 
  cost,
 
  -- product_line: 17 NULL rows → "Other"
  COALESCE(product_line, 'Other')    AS product_line,
 
  CAST(start_date AS DATE)          AS start_date
 
FROM workspace.bronze.products;
 
```

```python
%sql
-- Verifying that silver is Clean
SELECT
  COUNT(*) AS total_products,
  COUNT(category) AS has_category,
  COUNT(product_line) AS has_product_line
FROM workspace.silver.products;
```
